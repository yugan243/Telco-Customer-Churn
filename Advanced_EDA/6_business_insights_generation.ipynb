###  1. High-Risk Customer Profiles

Our analysis has identified two primary customer personas that are at an **exceptionally high risk of churning**. Understanding these profiles is the first step toward targeted intervention.

####  Profile A: The High-Risk "Trial" User
- **Churn Rate:** Over 53%
- **Behavior:** New customers (low tenure) using **Fiber optic** service
- **Commitment:** Very low ‚Äì **Month-to-month** contract + **Electronic check** payment
- **Engagement:** Minimal or no add-on services

>  These are our most volatile and highest-risk customers.

####  Profile B: The At-Risk "Basic" User
- **Churn Rate:** Over 26%
- **Behavior:** New customers (low tenure) using **DSL** internet
- **Commitment:** **Month-to-month** contract
- **Engagement:** No use of add-on services

> This group is also at high risk but uses a more basic plan.

###  2. Actionable Retention Opportunities

Our analysis not only identified **who is likely to churn**, but also **why**. This enables us to formulate the following data-driven strategies:

####  Recommendation 1: Focus on Contract Upgrades
- **Insight:** Contract type is the **strongest predictor** of loyalty.
- **Action:** 
  - Launch a promotional campaign for all **Month-to-month** customers.
  - Target new users (first 3‚Äì6 months).
  - Offer incentives like **1 free month** or **discounts** to upgrade to **One-year** contracts.

---

####  Recommendation 2: Cross-Sell "Core Protection" to Fiber Optic Users
- **Insight:** Churn among Fiber optic users drops significantly when they subscribe to **OnlineSecurity** or **TechSupport**.
- **Action:** 
  - Bundle or promote **"Core Protection" services** to all Fiber optic customers.
  - Position this as an upsell or loyalty benefit.

---

####  Recommendation 3: Incentivize "Sticky" Payment Methods
- **Insight:** **Electronic check** users churn more than those using automatic payments.
- **Action:** 
  - Offer a **one-time credit** or **monthly discount** for switching to:
    - **Bank transfer (automatic)** 
    - **Credit card (automatic)**
  - Reduce friction and increase customer commitment.

---

####  Recommendation 4: Nurture "Power Users"
- **Insight:** Highly engaged customers (‚Äú**Power Users**‚Äù) are the **most loyal**.
- **Action:** 
  - Create a **loyalty program** or "**Power User Track**"
  - Reward users for adopting **multiple services**.
  - Encourage long-term engagement with ecosystem-wide benefits.
